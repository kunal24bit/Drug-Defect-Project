```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
def vif_score(data):
    Scaler = StandardScaler()
    X = Scaler.fit_transform(data)
    vif_df = pd.DataFrame()
    vif_df["vif"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
    vif_df["features"] = data.columns
    return vif_df
    
    
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
def model_train_test(df, target, model):
    y = df[target]
    df1 = df.drop(target, axis = 1)
    x_train, x_test, y_train, y_test = train_test_split(df1,y, test_size = 0.24, random_state = 42)
    print("The shape is:\n")
    print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
    print("*****************************************\n")
    print("Training of the model is in progress...\n")
    model.fit(x_train, y_train)
    print("The R-square of the taken model is:\n")
    print(model.score(x_test, y_test))
    print("*****************************************")
    print("Score on the training set is:\n")
    print(model.score(x_train, y_train))
    print("\n***************************************\n")
    print("The prediction vector of the model is:\n")
    preds = model.predict(x_test)
    print(preds)
    print("******************************************")
    print("The mean squared error of the model is:\n")
    mse = mean_squared_error(preds, y_test)
    print(mse)
    
    
    
    
```

```python
model = RandomForestRegressor()
model_train_test(df5,"Defects", model)
```
